##Read Data From JSON 

```python
import json
fd=open("record.json",'r')
r=fd.read()
fd.close() 
```

```python
record=json.loads(r)
```

```python
record
```

##Add new Products to Inventory

```python
prod_id=str(input("Enter product id:"))

  
if prod_id in record.keys():
  add_quantity=int(input("Enter the quantity you want to add "))
  record[prod_id]['Avl.Quantity'] = record[prod_id]['Avl.Quantity'] + add_quantity
  print("Record Added")

else :
  print("New Record getting added,Please Enter the details")
  Name=str(input("Enter product name:"))
  category=str(input("Enter product category:"))
  Expiry_date=input("Enter Expiry date of Product:")
  Pr=int(input("Enter product price:"))
  Avl_quantity=int(input("Enter product Available Quantity:"))
  print("Product Added")
  record[prod_id]={'Name': Name,'Category':category,'Expiry Date':Expiry_date,'Price':Pr,'Avl.Quantity':Avl_quantity}

js=json.dumps(record)

fd=open("record.json",'w')
fd.write(js)
fd.close() 

    
```
